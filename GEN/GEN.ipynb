```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(123)
```

# Pierwsza funkcja do dopasowania

```python
def fit1(genome):
    return -(genome[0]**2+genome[1]**2+2*genome[2]**2)
    
```

# Druga funkcja do dopasowania

```python
def Rastrigin(genome):
    A=10
    n=len(genome)
    return -(A*n+np.sum(genome**2-A*np.cos(2*np.pi*genome)))
```

# domyślnie maksymalizujemy funkcje dopasowanie, dlatego powyższe funkcje są pomnożone przez -1

# algorytm genetyczny z mutacją gaussowską i krzyżowaniem jednopunktowym

Prawdopodobieństwo rodzicielstwa zależy od softmaxu funkcji dostosowania populacji. Będziemy mieli 100 generacji. I testowali dla rozmiarów populacji ze zbioru [10,24,50,100,200,300] i wariancji mutacji ze zbioru [0,0.25,0.5,0.75,1]. Rozmiar populacji w każdej generacji jest stały. Początkowy genotyp jest losowany jednostajnie z przedziału [-1,1].

```python
class GEN:
    
    class individual:
        def __init__(self,genome_size,var):
            self.genome=np.random.uniform(1,-1,genome_size)
            self.genome_size=genome_size
            self.var=var
            
        def mutate(self):
            self.genome[np.random.randint(self.genome_size)]+=np.random.normal(0,self.var)
    
    
    def __init__(self,population_size,genome_size,fit_function,var=0.25):
        self.fit_function=fit_function
        self.population=[self.individual(genome_size,var) for x in range(population_size)]
        self.genome_size=genome_size
        self.population_size=population_size
        self.var=var
    
    def softmax(self,z):
        e = np.exp(z)
        return e / np.sum(e, axis=0,keepdims=True)
    
    def cross(self,par1,par2):
        cut=np.random.randint(self.genome_size)
        chd1=self.individual(self.genome_size,self.var)
        chd2=self.individual(self.genome_size,self.var)
        chd1.genome[:cut]=par1.genome[:cut]
        chd1.genome[cut:]=par2.genome[cut:]
        chd2.genome[:cut]=par2.genome[:cut]
        chd2.genome[cut:]=par1.genome[cut:]
        return chd1,chd2
    
    
    def fitness(self):
        fit=np.zeros(self.population_size)
        for i in range(self.population_size):
            p=self.population[i]
            fit[i]=self.fit_function(p.genome)
            
        return fit
    
    def train(self,generations):
        fit=np.zeros(self.population_size)
        for g in range(generations):
            new_population=[0 for x in range(self.population_size)]
            for i in range(self.population_size):
                p=self.population[i]
                p.mutate()
                fit[i]=self.fit_function(p.genome)
                
            prob=self.softmax(fit)
            parents=np.random.choice(self.population,self.population_size,p=prob)
            for c in range(self.population_size//2):
                new_population[2*c], new_population[2*c+1] = self.cross(parents[2*c],parents[2*c+1])
                
            self.population=new_population
            
                
            
```

# fit1 $x^2+y^2+2z^2$

## generujemy statystyki podsumowywujące każdą generację w zależności od wariancji mutowania gausssowskiego i rozmiaru populacji

```python
df1=pd.DataFrame(columns=["max","min","median","variance","generation","mutation_var","population_size"])
for pop in [10,24,50,100,200,300]: 
    for v in np.arange(0,1.1,0.25):
        gen=GEN(pop,3,fit1,v)
        for x in range(101):
            temp=gen.fitness()
            ma=np.max(temp)
            mi=np.min(temp)
            me=np.median(temp)
            va=np.var(temp)
            df1=df1.append({"max":ma,"min":mi,"median":me,"variance":va,"generation":x,"mutation_var":v,"population_size":pop},ignore_index=True)
            gen.train(1)
```

## dostosowanie najlepszego osobnika w zależności od rozmiaru populacji i wariancji mutowania w ostatniej(100) generacji 

```python
sns.relplot(y="max",x="population_size",kind="line",data=df1[df1.generation==100],hue="mutation_var",palette=sns.color_palette("Set1", 5))
```

Jak widzimy dla małego rozmiar populacji wysoka wariancja daje bardzo złe rezultaty i samym mechanizmem krzyżowania też się nie da wiele osiągnąć, dla populacji większej od 24 rezultaty stają się bardziej sensowne

```python
sns.relplot(y="max",x="population_size",kind="line",data=df1.query("generation==100 and population_size>24"),hue="mutation_var",palette=sns.color_palette("Set1", 5))
```

najlepiej okazało się dać małą mutacyjność i dużą populację, co ciekawe nawet używając tylko mechanizmu krzyżowania można osiągnąć dobre rezultaty.

# fit2 Rastringa

## generujemy statystyki podsumowywujące każdą generację w zależności od wariancji mutowania gausssowskiego i rozmiaru populacji

```python
df2=pd.DataFrame(columns=["max","min","median","variance","generation","mutation_var","population_size"])
for pop in [10,24,50,100,200,300]: 
    for v in np.arange(0,1.1,0.25):
        gen=GEN(pop,5,Rastrigin,v)
        for x in range(101):
            temp=gen.fitness()
            ma=np.max(temp)
            mi=np.min(temp)
            me=np.median(temp)
            va=np.var(temp)
            df2=df2.append({"max":ma,"min":mi,"median":me,"variance":va,"generation":x,"mutation_var":v,"population_size":pop},ignore_index=True)
            gen.train(1)
```

## dostosowanie najlepszego osobnika w zależności od rozmiaru populacji i wariancji mutowania w ostatniej(100) generacji 

```python
sns.relplot(y="max",x="population_size",kind="line",data=df2[df2.generation==100],hue="mutation_var",palette=sns.color_palette("Set1", 5))
```

Bez mechanizmu mutacji nie da się osiągnąć dobrych rezultatów i też trzeba większej populacji

```python
sns.relplot(y="max",x="population_size",kind="line",data=df1.query("generation==100 and population_size>24 and mutation_var>0"),hue="mutation_var",palette=sns.color_palette("Set1", 4))
```

Ponownie najlepsza okazała się mała mutacyjność i duża populacja, trzeba jednak zaznaczyć że duża mutacyjność może na 3 krotnie mniejszej populacji osiągnąć zbliżone rezultaty

# Wnioski

Dobór mutacyjności i rozmiaru populacji jest kluczowy. By osiągnąć najdokładniejsze rezultaty najlepiej dać dużą populację i małą mutacyjność. Za mała populacja daję złe rezultaty dla każdej mutacyjności. Duża mutacyjność pozwala osiągnąć lepsze rezultaty przy mniejszej populacji niż w przypadku niskiej mutacyjności. Za duża populacja może spowalniać uczenie.
