```python
def fit1(rect,genome):
    return np.sum(rect[genome[:,2]==1,2])
```

# algorytm genetyczny z mutacją gaussowską i krzyżowaniem jednopunktowym

Prawdopodobieństwo rodzicielstwa zależy to stosunek dostosowania do sumy dostosowania całej populacji. Będziemy mieli 100 generacji. I testowali dla rozmiarów populacji 100 i wariancji mutacji 25. Rozmiar populacji w każdej generacji jest stały. Dokonujemy mutacji w ilości 1/4 wszystkich prostokątów dla danego osobnika. Pojedyńczy gen może być mutowany wielokrotnie. Początkowy genotyp jest losowany jednostajnie z przedziału [-R,R] gdzie R to rozmiar koła. Genotyp koduje położenie środka każdego prostokąta i jego aktywność. Jeśli prostokąt nie jest aktywny to nie liczy się do funkcji celu. Prostokąt może być aktywny tylko wtedy gdy nie pokrywa się z innymi aktywnymi prostokątami, ani nie wychodzi poza koło. 

Ponieważ prostokąt może być wybrany dowolną ilość razy to po prostu zbiór możliwych prostokątów zostanie zwielokrotniony. To znaczy że jak mamy na przykład zbiór 4 prostokątów dwukrotnie zwielokrotniony to by wybrać drugi prostokąt dwa razy to aktywujemy 3 i 4 prostokąt bo 4 jest jego duplikatem. Zwielokrotnie będzie równe stosunkowi pola koła do najmniejszego pola prostokąta dzielonemu przez 8 dlatego że jest mało prawdopodobne by tylko najmniejszy prostokąt był wykorzystywany i zapełnił cały okrąg i by zmniejszyć czas obliczeń.

Jak prostokąt wychodzi poza koło to się go daję z powrotem do koła ("odbija się"). Jak prostokąt aktywny zderzy się z innym prostokątem aktywnym to staje się nieaktywny i ustawia się na styk do prostokąta z którym kolidował.
Legalność akywności i sprawdzanie czy coś może stać się aktywne jest sprawdzane w losowej kolejności.

```python
class CUT:
    
    class individual:
        def __init__(self,genome_size,var,R):
            self.genome=np.random.uniform(R,-R,[genome_size,3])
            self.genome[:,2]=(self.genome[:,2]>0).astype(float)
            self.genome_size=genome_size
            self.var=var
            
        def mutate(self):
            for i in range(self.genome_size//4):
                self.genome[np.random.randint(self.genome_size),0]+=np.random.normal(0,self.var)
                self.genome[np.random.randint(self.genome_size),1]+=np.random.normal(0,self.var)
            
            
        def is_legal(self,rect,R,verbose=False):
            
            legal=self.genome[:,2]==1
            for i in np.random.permutation(np.argwhere(legal)):
                #check circle
                dx=max(abs(self.genome[i,0]+rect[i,0]/2),abs(self.genome[i,0]-rect[i,0]/2))
                dy=max(abs(self.genome[i,1]+rect[i,1]/2),abs(self.genome[i,1]-rect[i,1]/2))
                if R**2<=dx**2+dy**2:
                    #rebounce
                    if verbose:
                        print(str(i)+" not in circle")
                    if (dx**2+dy**2)>0.00001:
                        self.genome[i,0]=self.genome[i,0]*(0.9*R**2/(dx**2+dy**2))
                        self.genome[i,1]=self.genome[i,1]*(0.9*R**2/(dx**2+dy**2))
                    self.genome[i,2]=0
                    continue
                
                #check rect
                for j in np.random.permutation(np.argwhere(self.genome[:,2]==1)):
                    if i!=j:
                        if abs(self.genome[i,0]-self.genome[j,0])<rect[i,0]/2+rect[j,0]/2 and abs(self.genome[i,1]-self.genome[j,1])<rect[i,1]/2+rect[j,1]/2:
                            #rebounce
                            if abs(self.genome[i,0]-self.genome[j,0])-(rect[i,0]/2+rect[j,0]/2)<abs(self.genome[i,1]-self.genome[j,1])-(rect[i,1]/2+rect[j,1]/2):
                                if self.genome[i,0]<self.genome[j,0]:
                                    self.genome[i,0]=self.genome[i,0]-(abs(self.genome[i,0]-self.genome[j,0])-(rect[i,0]/2+rect[j,0]/2))-0.01
                                else:
                                    self.genome[i,0]=self.genome[i,0]+(abs(self.genome[i,0]-self.genome[j,0])-(rect[i,0]/2+rect[j,0]/2))+0.01
                            else:
                                if self.genome[i,1]<self.genome[j,1]:
                                    self.genome[i,1]=self.genome[i,1]-(abs(self.genome[i,1]-self.genome[j,1])-(rect[i,1]/2+rect[j,1]/2))-0.01
                                else:
                                    self.genome[i,1]=self.genome[i,1]+(abs(self.genome[i,1]-self.genome[j,1])-(rect[i,1]/2+rect[j,1]/2))+0.01
                            
                            if verbose:
                                print(str(i)+" collision with "+str(j))
                            
                            self.genome[i,2]=0
                            break
                    
        def make_legal(self,rect,R,verbose=False):
            
            illegal=self.genome[:,2]==0
            for i in np.random.permutation(np.argwhere(illegal)):
                self.genome[i,2]=1
                dx=max(abs(self.genome[i,0]+rect[i,0]/2),abs(self.genome[i,0]-rect[i,0]/2))
                dy=max(abs(self.genome[i,1]+rect[i,1]/2),abs(self.genome[i,1]-rect[i,1]/2))
                if R**2<=dx**2+dy**2:
                    #rebounce
                    if verbose:
                        print(str(i)+" not in circle")
                    if (dx**2+dy**2)>0.00001:
                        self.genome[i,0]=self.genome[i,0]*(0.9*R**2/(dx**2+dy**2))
                        self.genome[i,1]=self.genome[i,1]*(0.9*R**2/(dx**2+dy**2))
                    self.genome[i,2]=0
                    continue
                
                #check rect
                for j in np.random.permutation(np.argwhere(self.genome[:,2]==1)):
                    if i!=j:
                        if abs(self.genome[i,0]-self.genome[j,0])<rect[i,0]/2+rect[j,0]/2 and abs(self.genome[i,1]-self.genome[j,1])<rect[i,1]/2+rect[j,1]/2:
                            if verbose:
                                print(str(i)+" collision with "+str(j))
                            
                            self.genome[i,2]=0
                            break
                    
        
        
    
    def __init__(self,population_size,genome_size,fit_function,R,rect,var=0.25):
        self.fit_function=fit_function
        self.population=[self.individual(genome_size,var,R) for x in range(population_size)]
        self.genome_size=genome_size
        self.population_size=population_size
        self.var=var
        self.R=R
        self.rect=rect
    
    def normalize(self,z):
        return z / np.sum(z, axis=0,keepdims=True)
    
    def cross(self,par1,par2):
        cut=np.random.randint(self.genome_size)
        chd1=self.individual(self.genome_size,self.var,R=1)
        chd2=self.individual(self.genome_size,self.var,R=1)
        chd1.genome[:cut]=par1.genome[:cut]
        chd1.genome[cut:]=par2.genome[cut:]
        chd2.genome[:cut]=par2.genome[:cut]
        chd2.genome[cut:]=par1.genome[cut:]
        return chd1,chd2

    
    def fitness(self):
        fit=np.zeros(self.population_size)
        for i in range(self.population_size):
            p=self.population[i]
            fit[i]=self.fit_function(self.rect,p.genome)
            
        return fit
    
    def train(self,generations):
        fit=np.zeros(self.population_size)
        for g in range(generations):
            new_population=[0 for x in range(self.population_size)]
            for i in range(self.population_size):
                p=self.population[i]
                p.mutate()
                p.is_legal(self.rect,self.R)
                p.make_legal(self.rect,self.R)
                fit[i]=self.fit_function(self.rect,p.genome)
            
            
            prob=self.normalize(fit)
            parents=np.random.choice(self.population,self.population_size,p=prob)
            for c in range(self.population_size//2):
                new_population[2*c], new_population[2*c+1] = self.cross(parents[2*c],parents[2*c+1])
                
            self.population=new_population
        
        for i in range(self.population_size):
            p=self.population[i]
            p.is_legal(self.rect,self.R)
            p.make_legal(self.rect,self.R)
        
                
            
```

## Początkowe ustawienia: można zwiększać POP i GEN, zmniejszać MUL by osiągać lepsze rezultaty

populacja startowa

```python
POP=100
```

ilość generacji

```python
GEN=50
```

pomniejszenie ilości powtórzeń

```python
MUL=8
```

# r1000

# Active

```python
#load
rect=np.genfromtxt("cutting/r1000.csv", delimiter=',')

#multiply
rect=np.repeat(rect,int(np.pi*1000*1000/min(rect[:,0]*rect[:,1])/MUL),axis=0)

cut1000=CUT(population_size=POP,genome_size=rect.shape[0],fit_function=fit1,R=1000,rect=rect,var=25)

#train
cut1000.train(GEN)

#visualize
best=cut1000.population[np.argmax(cut1000.fitness())].genome
X = best[best[:,2]==1,0]
Y = best[best[:,2]==1,1]
R=rect[best[:,2]==1,:]

fig,ax = plt.subplots(1,figsize=(10, 10))
ax.add_patch(pt.Circle((0,0),1000))
for i in range(X.shape[0]):
    ax.add_patch(pt.Rectangle((X[i]-R[i,0]/2, Y[i]-R[i,1]/2),R[i,0], R[i,1], facecolor='r'))
    plt.text(X[i]-rect[i,0]/4,Y[i]-R[i,1]/4,str(int(R[i,2])))
    
ax.set_xlim(-1300, 1300)
ax.set_ylim(-1300, 1300)
```

Jak widzimy rozwiązanie jest niesatysfakcjonujące. Mamy duże puste obszary.

```python
#visualize
best=cut1000.population[np.argmax(cut1000.fitness())].genome
X = best[best[:,2]!=1,0]
Y = best[best[:,2]!=1,1]
R=rect[best[:,2]!=1,:]

fig,ax = plt.subplots(1,figsize=(10, 10))
ax.add_patch(pt.Circle((0,0),1000))
for i in range(X.shape[0]):
    ax.add_patch(pt.Rectangle((X[i]-R[i,0]/2, Y[i]-R[i,1]/2),R[i,0], R[i,1], facecolor='r'))
    plt.text(X[i]-rect[i,0]/4,Y[i]-R[i,1]/4,str(int(R[i,2])))
    
ax.set_xlim(-1300, 1300)
ax.set_ylim(-1300, 1300)
```

Widzymy że nasz podstawowy algorytm dla populacji 100 i 50 generacjach nie radzi sobie dobrze. Możliwe że to dlatego że po prostu parametry modeli były za małe. Po za tym są problemy z przygotowywaniem się do położenia dużych prostokątów. Model też nie patrzył na stosunek wartości do pola ani wartość podczas kolizji. Możliwe też że wariancja była za mała.
